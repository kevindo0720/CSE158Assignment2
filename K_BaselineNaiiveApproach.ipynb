```python
import csv
from collections import defaultdict

csv_file_path = './archive/interactions_train.csv'

reviews = defaultdict(list)\
    
reviewsRows = []
with open(csv_file_path, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row[0] == "user_id":
            continue
        else:
            reviewsRows.append(row)

# check if reviews populated
DictUI = {}
DictIU = {}
DictInteraction = {}


#populate all of the dictionary
for i in reviewsRows:
    if i[0] not in DictUI.keys():
        DictUI[i[0]] = [i[1]]
    else:
        DictUI[i[0]].append(i[1])
        
    if i[1] not in DictIU.keys():
        DictIU[i[1]] = [i[0]]
    else:
        DictIU[i[1]].append(i[0])
        
    
    DictInteraction[i[0],i[1]] = float(i[3])
    


        
```

```python
#find the average of all the reviews for each user
DictURatings = {}

for user in DictUI.keys():
    
    listOfRecipes = DictUI[user]
    allReviewsForCurrUser = []
    
    for recipe in listOfRecipes:
        allReviewsForCurrUser.append(DictInteraction[user,recipe])
    averageCurrUser = sum(allReviewsForCurrUser)/len(allReviewsForCurrUser)
    
    DictURatings[user] = averageCurrUser
    

print(DictURatings)

#global average 
globalAvg = []
for u in DictURatings.keys():
    globalAvg.append(DictURatings[u])
    
globalAvgVal = sum(globalAvg)/len(globalAvg)



         
```

```python
#get test set

csv_file_path_test = './archive/interactions_test.csv'

    
reviewsRowstest = []
with open(csv_file_path_test, 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row[0] == "user_id":
            continue
        else:
            reviewsRowstest.append(row)

# check if reviews populated

Interactiontest = []
InteractionpredictAvg = []
InteractionpredictFives = []


#populate all of the dictionary
for i in reviewsRowstest:
    if i[0] in DictURatings.keys():
        InteractionpredictAvg.append(DictURatings[i[0]])
    else:
        InteractionpredictAvg.append(globalAvgVal)
        
    InteractionpredictFives.append(5.0)
        
    Interactiontest.append(float(i[3]))
    

print(InteractionpredictAvg)
print(Interactiontest)
print(InteractionpredictFives)


    
```

```python
def MSE(predictions, labels):
    differences = [(x-y)**2 for x,y in zip(predictions,labels)]
    return sum(differences) / len(differences)

msePredictNaiiveAverage = MSE(InteractionpredictAvg,Interactiontest)
msePredictNaiiveFives = MSE(InteractionpredictFives,Interactiontest)

print("Baseline - users average approach: " + str(msePredictNaiiveAverage))
print("Baseline - rate every single review as a 5: " + str(msePredictNaiiveFives) )
```
